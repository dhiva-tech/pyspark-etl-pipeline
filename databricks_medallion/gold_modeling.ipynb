```python
df = spark.table("silver_products")
```

```python
dim_product = df.select(
    "id",
    "brand",
    "category",
    "width",
    "height",
    "depth"
).dropDuplicates()
```

```python
fact_product = df.select(
    "id",
    "price",
    "discount_pct",
    "availability_status",
    "rating",
    "image_count"
)
```

```python
dim_product.write.format("delta").mode("overwrite").saveAsTable("dim_product")

fact_product.write.format("delta").mode("overwrite").saveAsTable("fact_product")
```
