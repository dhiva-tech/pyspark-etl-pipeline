```python
df=spark.table('bronze_products')
```

```python
from pyspark.sql.functions import col, size

df_silver = df.select(
    col("id"),
    col("brand"),
    col("category"),
    col("price"),
    col("discount_pct"),
    col("availabilityStatus"),
    col("rating"),
    size("images").alias("image_count"),
    col("dimensions")["width"].alias("width"),
    col("dimensions")["height"].alias("height"),
    col("dimensions")["depth"].alias("depth")
)
```

```python
df_silver = df_silver.filter(col("id").isNotNull())
df_silver = df_silver.dropDuplicates()
```

```python
df_silver = df_silver.withColumnRenamed("availabilityStatus", "availability_status")
```

```python
df_silver.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("silver_products")
```
